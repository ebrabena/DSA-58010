```python
import mysql.connector

mydb = mysql.connector.connect(
    host = "localhost",
    user = "root",
    passwd = "myPhone",
    database = "Phonebook"
)

mycursor = mydb.cursor()

a = 0
cancel = 1
condition1 = 0
user_action1 = (input("Type 1 for login, 2 for register.\n"))

if user_action1 == '1':
    select_local = "SELECT * FROM local_accounts WHERE name = %s"
    name_temp = input("Enter username: ")
    mycursor.execute(select_local, (name_temp, ))
    local_row = mycursor.fetchall()
    for result in local_row:
        password_temp = input("Enter password: ")
        temp_add = (name_temp, password_temp, 0)
        mydb.commit()
        if result == temp_add:
            print("Login successful!")
            condition1 = 1
            delete_temp = "DELETE FROM temp"
            mycursor.execute(delete_temp)
            mydb.commit()
        else:
            print("Login failed!")
            condition1 = 0
            delete_temp = "DELETE FROM temp"
            mycursor.execute(delete_temp)
            mydb.commit()

elif user_action1 == '2':
    sqlFormula = "INSERT INTO local_accounts (name, password, level_auth) VALUES (%s, %s, %s)"
    local_account_add = (input("Enter username: "), input("Enter password: "), 0)
    mycursor.execute(sqlFormula, local_account_add)
    mydb.commit()
    print("Registered successfully!")

elif user_action1 == '3':
    mycursor.execute("CREATE TABLE foreign_accounts (name VARCHAR(255), mark INTEGER(1))")

while a < 1:
    b = 0
    c = 0
    num = 0
    if condition1 == 1:
        user_action2 = (input("\nType\n1 to view contacts\n2 to view messages\n3 to create emails\n4 to end session\n\n"))

        if user_action2 == '1':
            contact_view = input("\nType\n1 to view local contacts\n2 to view foreign contacts\n3 to cancel\n\n")
            while b < 1:
                if contact_view == '1':
                    mycursor.execute("SELECT name, level_auth FROM local_accounts")
                    view_accounts_local = mycursor.fetchall()
                    for view_local in view_accounts_local:
                        num += 1
                        contacts = []
                        contacts.append(num)
                        contacts.append(view_local)
                        print(contacts)
                    else:
                        print("Reached the end of list.")
                        b += cancel
                elif contact_view == '2':
                    mycursor.execute("SELECT * FROM foreign_accounts")
                    view_accounts_foreign = mycursor.fetchall()
                    for view_foreign in view_accounts_foreign:
                        num += 1
                        contacts = []
                        contacts.append(num)
                        contacts.append(view_foreign)
                        print(contacts)
                    else:
                        print("There are currently no foreign contacts saved!")
                        b += cancel
                elif contact_view == '3':
                    b += cancel

        elif user_action2 == '2':
            which_view = input("\nType\n1 to view inbox\n2 to view sent messages\n3 to view drafts\n4 to cancel\n\n")
            while b < 1:
                if which_view == '1':
                    mycursor.execute("SELECT msg_to, msg_from, msg FROM msg_received")
                    received_msg = mycursor.fetchall()
                    for received in received_msg:
                        num += 1
                        msgs = []
                        msgs.append(num)
                        msgs.append(received)
                        print(msgs)
                    else:
                        print("Reached the end of list.")
                        b += cancel
                elif which_view == '2':
                    mycursor.execute("SELECT msg_to, msg_from, msg FROM msg_sent")
                    sent_msg = mycursor.fetchall()
                    for sent in sent_msg:
                        num += 1
                        msgs = []
                        msgs.append(num)
                        msgs.append(sent)
                        print(msgs)
                    else:
                        print("Reached the end of list.")
                        b += cancel
                elif which_view == '3':
                    mycursor.execute("SELECT * FROM drafts")
                    drafts_msg = mycursor.fetchall()
                    for drafts in drafts_msg:
                        num += 1
                        msgs = []
                        msgs.append(num)
                        msgs.append(drafts)
                        print(msgs)
                    else:
                        print("Reached the end of list.")
                        b += cancel
                elif which_view == '4':
                    b += cancel

        elif user_action2 == '3':
            while b < 1:
                c = 0
                msg = input("\nType a message. Type 'CANCEL' to cancel.\n\n")
                if msg == 'CANCEL':
                    b += 1
                else:
                    willSend = input("Confirm to send? Y/N\n")
                    if willSend == 'Y':
                        while c < 1:
                            send_to = input("Send to single or multiple recipient/s?\nType\n1 for single\n2 for multiple\n3 to cancel\n\n")
                            if send_to == '1':
                                select_recipient = input("Type the username of your recipient: ")
                                sqlFormula = "INSERT INTO msg_sent (msg_to, msg_from, msg, mark) VALUES (%s, %s, %s, %s)"
                                send_msg = (select_recipient, name_temp, msg, 0)
                                mycursor.execute(sqlFormula, send_msg)
                                mydb.commit()
                                print("Message sent successfully!")
                                c += 1
                            elif send_to == '2':
                                num_recipient = int(input("Type the amount of recipients: "))
                                i = 0
                                while i < num_recipient:
                                    select_recipient = input("Type the username of your recipient: ")
                                    sqlFormula = "INSERT INTO msg_sent (msg_to, msg_from, msg, mark) VALUES (%s, %s, %s, %s)"
                                    send_msg = (select_recipient, name_temp, msg, 0)
                                    mycursor.execute(sqlFormula, send_msg)
                                    mydb.commit()
                                    i += 1
                                else:
                                    print("Messages sent successfully!")
                                    c += 1
                            else:
                                c += 1
                                print("Action aborted.")
                    elif willSend == 'N':
                        sqlFormula = "INSERT INTO drafts (msg_from, msg) VALUES (%s, %s)"
                        drafts_add = (name_temp, msg)
                        mycursor.execute(sqlFormula, drafts_add)
                        mydb.commit()
                        print("Message saved as draft!")
                        b += cancel

        else:
            a += cancel
            print("Session ended.\nLogging out.")
```
